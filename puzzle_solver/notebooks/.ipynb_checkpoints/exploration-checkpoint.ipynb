## 0. Setup

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider, Checkbox, Button, HBox, VBox, Output
import os

%matplotlib inline

def show_image(img, title="Image", cmap=None):
    plt.figure(figsize=(10, 8))
    if len(img.shape) == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap=cmap or 'gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Load Image
image_path = '../data/pieces.jpg'
if os.path.exists(image_path):
    image = cv2.imread(image_path)
    print(f"Loaded image from {image_path}")
else:
    print("Image not found! Using dummy image.")
    image = np.zeros((500, 500, 3), dtype=np.uint8)
    cv2.circle(image, (250, 250), 100, (255, 255, 255), -1)

show_image(image, "Original")
```

## 1. HSV Thresholding
Adjust sliders to isolate the puzzle pieces (or background).
Use `Show Mask` to see the binary output.

```python
def hsv_pipeline(h_min, h_max, s_min, s_max, v_min, v_max, blur_size, morph_open, morph_close, show_mask):
    # 1. Blur
    if blur_size > 0:
        k = blur_size * 2 + 1
        blurred = cv2.GaussianBlur(image, (k, k), 0)
    else:
        blurred = image.copy()
        
    # 2. HSV
    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)
    lower = np.array([h_min, s_min, v_min])
    upper = np.array([h_max, s_max, v_max])
    mask = cv2.inRange(hsv, lower, upper)
    
    # 3. Morphology
    kernel = np.ones((3,3), np.uint8)
    if morph_open > 0:
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=morph_open)
    if morph_close > 0:
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=morph_close)
        
    # 4. Results
    print(f"Configuration: Lower=[{h_min}, {s_min}, {v_min}], Upper=[{h_max}, {s_max}, {v_max}]")
    
    if show_mask:
        show_image(mask, "Binary Mask", cmap='gray')
    else:
        res = cv2.bitwise_and(image, image, mask=mask)
        show_image(res, "Segmented")

interact(hsv_pipeline,
         h_min=IntSlider(min=0, max=179, value=0),
         h_max=IntSlider(min=0, max=179, value=179),
         s_min=IntSlider(min=0, max=255, value=0),
         s_max=IntSlider(min=0, max=255, value=255),
         v_min=IntSlider(min=0, max=255, value=0),
         v_max=IntSlider(min=0, max=255, value=255),
         blur_size=IntSlider(min=0, max=10, value=2),
         morph_open=IntSlider(min=0, max=5, value=1),
         morph_close=IntSlider(min=0, max=5, value=1),
         show_mask=Checkbox(value=False))
```

## 2. Adaptive Thresholding
Useful if lighting is uneven.

```python
def adaptive_pipeline(block_size, C, invert):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    bs = block_size if block_size % 2 == 1 else block_size + 1
    type_ = cv2.THRESH_BINARY_INV if invert else cv2.THRESH_BINARY
    
    thresh = cv2.adaptiveThreshold(blurred, 255, 
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                   type_, bs, C)
    
    print(f"Config: BlockSize={bs}, C={C}, Invert={invert}")
    show_image(thresh, "Adaptive Threshold", cmap='gray')

interact(adaptive_pipeline,
         block_size=IntSlider(min=3, max=51, step=2, value=11),
         C=IntSlider(min=-10, max=20, value=2),
         invert=Checkbox(value=False))
```

## 3. Contour Detection & Filtering
Visualize how area filtering affects which pieces are kept.

```python
def contour_pipeline(min_area, max_area):
    # Standard preprocessing for this demo
    # You can change this to use values found in step 1
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Simple Otsu for demonstration
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    filtered_contours = [c for c in contours if min_area < cv2.contourArea(c) < max_area]
    
    display = image.copy()
    cv2.drawContours(display, filtered_contours, -1, (0, 255, 0), 2)
    
    print(f"Found {len(filtered_contours)} pieces (from {len(contours)} total).")
    show_image(display, "Contours")

interact(contour_pipeline,
         min_area=IntSlider(min=0, max=5000, value=1000),
         max_area=IntSlider(min=5000, max=100000, value=50000))
```
